```python
from graphviz import Digraph

# Define the main idea or problem statement
main_idea = "Negative Customer Experience at Whole Foods Market"

# Define the key drivers and outcomes
key_drivers = ["Inadequate Staff Assistance", "Limited Parking Space", "Long Checkout Lines", "Crowded Store Aisles", "Inefficient Store Navigation", "Poor Store Layout"]
key_outcomes = ["Poor Customer Service", "Negative Customer Reviews", "Decreased Customer Loyalty", "Decreased Sales Revenue"]

# Define the relationships between the key drivers, outcomes, and main idea
relationships = [("Inadequate Staff Assistance", "Negative Customer Experience at Whole Foods Market"),
                 ("Limited Parking Space", "Negative Customer Experience at Whole Foods Market"),
                 ("Long Checkout Lines", "Negative Customer Experience at Whole Foods Market"),
                 ("Crowded Store Aisles", "Negative Customer Experience at Whole Foods Market"),
                 ("Inefficient Store Navigation", "Negative Customer Experience at Whole Foods Market"),
                 ("Poor Store Layout", "Negative Customer Experience at Whole Foods Market"),
                 ("Inadequate Staff Assistance", "Poor Customer Service"),
                 ("Poor Customer Service", "Negative Customer Reviews"),
                 ("Negative Customer Reviews", "Decreased Customer Loyalty"),
                 ("Decreased Customer Loyalty", "Decreased Sales Revenue")]

# Create a new Digraph
dot = Digraph()
dot.graph_attr['size'] = '25,20'
# Set graph properties
dot.graph_attr['rankdir'] = 'LR'
dot.node_attr['shape'] = 'rectangle'

# Add main idea node
dot.node(main_idea, main_idea, shape='diamond')

# Add key driver nodes
for driver in key_drivers:
    dot.node(driver, driver)

# Add key outcome nodes
for outcome in key_outcomes:
    dot.node(outcome, outcome)

# Add relationships between nodes
for driver, outcome in relationships:
    dot.edge(driver, outcome)

# Display the graph
dot.render('internal_diagram', format='jpeg')
display(dot)
```
